#### Question_3 SET

```
S={1,4,5,8,3,6}
X={3,4,9,10,11}
print(S.union(X))
```

```
print(S.intersection(X))
```

```
print(S.difference(X))
```

```
print(S.symmetric_difference(X))
```

```
X.discard(11)
print(X)
```
